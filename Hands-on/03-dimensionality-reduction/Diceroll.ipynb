Let's create data coming from a boardgame player rolling two dice, adding them up, and then making a move based on the result.

```python
import numpy as np
import seaborn as sns
import plotly.express as px
from sklearn.decomposition import PCA
import pandas as pd
```

```python
# Step 1: Create dice roll variables
X = np.random.randint(1, 7, 1000)
Y = np.random.randint(1, 7, 1000)
Z = X+Y

df = pd.DataFrame({'first die': X, 'second die': Y, 'result': Z})
```

Standardize the attributes.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
scaler = StandardScaler()
df = pd.DataFrame(scaler.fit_transform(df), columns = df.columns)

df
```

Plot

```python
fig = px.scatter_3d(df, x='first die', y='second die', z='result', title='3D Scatter plot of Dice Rolls')
fig.show()
```

Can we do dimensionality reduction without losing any information?

```python
# Step 4: Perform PCA
pca = PCA()
df_pca = pca.fit_transform(df)
```

```python
# Step 5: Scree plot using Seaborn
fig = sns.barplot(x=np.arange(1, pca.n_components_ + 1), y=pca.explained_variance_ratio_)

print(pca.explained_variance_ratio_)
```
